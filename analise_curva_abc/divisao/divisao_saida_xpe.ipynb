# Divisão de Saída Flowrack (XPE)

## Import

```python
import pandas as pd
pd.set_option('display.max_columns', None)

import math as mt
```

## Func

```python
def refatorar_indece(df, nome_index):
    
    qnt_linha = df.shape[0] + 1

    index = [i for i in range(1, qnt_linha)]

    df.set_index(pd.Index(index), inplace=True)

    df.index.name = nome_index
    
    return df
```

### Seleção

```python
adicionar_cod_xpe = [10411, 10410, 10122, 12217, 101051, 232465]
retirar_cod_n_xpe = [231225, 922092, 130042]

selecao_add_cod_xpe =   (curva_frac['Código'].isin(adicionar_cod_xpe)) & \
                        (curva_frac['Curva Frac'] == 'A')
selecao_retirar_cod_xpe = curva_frac['Código'].isin(retirar_cod_n_xpe)

curva_frac[selecao_retirar_cod_xpe]
```

```python
 # curva A que contenha [emb = 60 e curva = A e (dec = xpe|susp|sol|elixir ou cod = selecionado previamente) e desc != 30 ml|30ml| 20ml|/20ml|2,5ml|15ml]
selecao_desc =  (curva_frac['Embal.'] == 60.0) & \
                (curva_frac['Curva Frac'] == 'A') & \
                (curva_frac['Descrição'].str.contains('xpe|susp|sol|elixir', case=False)) | \
                (selecao_add_cod_xpe) & \
                (~curva_frac['Descrição'].str.contains('30 ml|30ml| 30ml| 20ml|/20ml|2,5ml|15ml', case=False))
```

```python
xpe = curva_frac[selecao_desc]
xpe
```

```python
xpe.drop(xpe[selecao_retirar_cod_xpe].index, inplace = True)
xpe.to_excel('teste.xlsx')
```

### Divisão

```python
refatorar_indece(xpe, 'Ordem')

cod_xpe = xpe['Código']

desc_xpe = xpe['Descrição']

saida_xpe = xpe['Qtde Venda Frac']
```

```python
qnt_enderecos_modulo_xpe = 9
print((f'Total de endereços de xarope por modulo: {qnt_enderecos_modulo_xpe}'))

modulos = 6
print(f'Total de modulos: {modulos}')

total_enderecos_xpe = qnt_enderecos_modulo_xpe * modulos
print(f'Total de endereços xarope: {total_enderecos_xpe}')

total_curva_a_xpe = xpe.shape[0]
print(f'Total de xaropes: {total_curva_a_xpe}')

qnt_itens_modulo_xpe = mt.ceil(total_curva_a_xpe / modulos)
print(f'Total de xaropes por modulo: {qnt_itens_modulo_xpe}')

ordem_modulo_xpe = [x for x in range(1, qnt_enderecos_modulo_xpe + 1)]
ordem_modulo_xpe
```

```python
step = 6
```

```python
# Ordem e saida dos itens no flowrack (XAROPE)

primeiro_m = [x for x in range(1, qnt_enderecos_modulo_xpe * step, step)]

segundo_m = [x for x in range(3, qnt_enderecos_modulo_xpe * step, step)]

terceiro_m = [x for x in range(5, qnt_enderecos_modulo_xpe * step, step)]

quarto_m = [x for x in range(6, qnt_enderecos_modulo_xpe * step + 1, step)]

quinto_m = [x for x in range(4, qnt_enderecos_modulo_xpe * step, step)]

sexto_m = [x for x in range(2, qnt_enderecos_modulo_xpe * step, step)]
```

```python
modulos = [{'29_28' : primeiro_m},
           {'27_26' : segundo_m},
           {'25_24' : terceiro_m},
           {'23_22' : quarto_m},
           {'21_20' : quinto_m},
           {'19_18' : sexto_m},
          ]
modulos
```

```python
# Criar um DataFrame vazio
divisao = pd.DataFrame()

# Iterar sobre os dicionários da lista
for d in modulos:
    # Iterar sobre as chaves e listas do dicionário
    for col_name, values in d.items():
        # Adicionar as células com os valores da lista na coluna correspondente
        divisao[col_name] = values

refatorar_indece(divisao, 'Ordem')
divisao
```

```python
div_desc = divisao.replace(desc_xpe)

refatorar_indece(div_desc, 'Ordem')
div_desc.to_excel('analises_prontas\ordem\divisão_por_desc_xpe.xlsx')

div_desc
```

```python
div_cod = divisao.replace(cod_xpe)

refatorar_indece(div_cod, 'Ordem')
div_cod.to_excel('analises_prontas\ordem\divisão_por_cod_xpe.xlsx')

div_cod
```

```python
saida_xpe = divisao.replace(saida_xpe)

refatorar_indece(saida_xpe, 'Ordem')
div_cod.to_excel('analises_prontas\ordem\saida_xpe.xlsx')

saida_xpe
```

```python
junto = pd.concat([div_desc, div_cod, divisao, saida_xpe])
junto.to_excel('analises_prontas\ordem\cod_mais_desc_xpe.xlsx')
junto
```
