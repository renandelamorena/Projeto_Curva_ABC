```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
todos_classificar = pd.read_excel(r'C:\Users\estoque\Documents\estoque.renan\projetos\projeto_curva_abc\analise_curva_abc\divisao\itens_para_dividir_classificar.xlsx')
situacao_final = pd.read_excel(r'C:\Users\estoque\Documents\estoque.renan\projetos\projeto_curva_abc\tratamento_curva_abc\dados_tratados\situacao_final.xlsx')
```

```python
def refatorar_indece(df, nome_index):
    
    qnt_linha = df.shape[0] + 1

    index = [i for i in range(1, qnt_linha)]

    df.set_index(pd.Index(index), inplace=True)

    df.index.name = nome_index
    
    return df
```

```python
end_atuais = situacao_final[['Código', 'Ender.Fracionado']]
end_atuais['Ender.Fracionado'] = end_atuais['Ender.Fracionado'].astype(str)
```

```python
index_AA = ['AA' for i in range(27)]
index_AB = ['AB' for i in range(32)]
index_AC = ['AC' for i in range(20)]

index = (index_AA + index_AB + index_AC)
```

```python
enderecos_x_locais = pd.read_excel(r'C:\Users\estoque\Documents\estoque.renan\projetos\projeto_curva_abc\analise_curva_abc\local\locais_frac_classificados.xlsx')
enderecos_x_locais.set_index('Unnamed: 0', inplace = True)
enderecos_x_locais = enderecos_x_locais.astype(str)
```

# Modulo 28_29

### Verificando os itens a serem classificados

```python
# Transforma a primeira coluna em uma data frame diferente
modulo_29_28 = todos_classificar[['29_28']]

# Junta o df da divisão, com o df de endereços atuais, ficando o df da divisão com os endereços de fracionados atuais
modulo_29_28 = pd.merge(modulo_29_28, end_atuais, left_on = '29_28', right_on = 'Código')
modulo_29_28.drop_duplicates(subset = ['29_28'], inplace = True)

refatorar_indece(modulo_29_28, 'Ordem_modulo')

# Verifica se esta no modulo correto
modulo_29_28['Ender.Fracionado'] = modulo_29_28['Ender.Fracionado'].astype(str)
modulo_29_28['Modulo correto'] = modulo_29_28['Ender.Fracionado'].str.startswith(('29', '28'))

# Adicionado columa com as classes
modulo_29_28['Classes'] = index

# Renomear colunas
novos_nomes = {'Ender.Fracionado' : 'Endereço Atual',
               'Código' : 'Itens do modulo 29_28',
              }
modulo_29_28.rename(columns = novos_nomes, inplace = True)

# Endereço atual do item a classificar
classe_aa_29_28_classificar = modulo_29_28[modulo_29_28['Classes'] == 'AA']['Endereço Atual']
classe_ab_29_28_classificar = modulo_29_28[modulo_29_28['Classes'] == 'AB']['Endereço Atual']
classe_ac_29_28_classificar = modulo_29_28[modulo_29_28['Classes'] == 'AC']['Endereço Atual']

# Endereços das classes 
end_classe_aa_29_28 = enderecos_x_locais[enderecos_x_locais['Classe'] == 'AA']['29_28']
end_classe_ab_29_28 = enderecos_x_locais[enderecos_x_locais['Classe'] == 'AB']['29_28']
end_classe_ac_29_28 = enderecos_x_locais[enderecos_x_locais['Classe'] == 'AC']['29_28']

# Verificando se os itens a classificar foram classificados, na classe certa e no local certo.
item_x_class_aa_29_28 = classe_aa_29_28_classificar.isin(end_classe_aa_29_28)
item_x_class_ab_29_28 = classe_ab_29_28_classificar.isin(end_classe_ab_29_28)
item_x_class_ac_29_28 = classe_ac_29_28_classificar.isin(end_classe_ac_29_28)

nova_item_x_classes_29_28 = pd.concat([item_x_class_aa_29_28, item_x_class_ab_29_28, item_x_class_ac_29_28])

modulo_29_28['Classificado'] = nova_item_x_classes_29_28

modulo_29_28.drop(columns = '29_28', inplace = True)
modulo_29_28
```

```python
modulo_29_28_end = enderecos_x_locais[['29_28']]
modulo_29_28_end.set_index(pd.Index(index), inplace=True)

end_atuais['Código'] = end_atuais['Código'].astype(str)

modulo_29_28_end_x_cod = modulo_29_28_end.merge(end_atuais, how = 'left' , left_on = '29_28', right_on = 'Ender.Fracionado')

modulo_29_28_end_x_cod.drop_duplicates(inplace = True)
modulo_29_28_end_x_cod.drop(columns = 'Ender.Fracionado', inplace = True)
modulo_29_28_end_x_cod.fillna('-', inplace = True)

modulo_29_28_end_x_cod.set_index(pd.Index(index), inplace=True)

modulo_29_28_end_x_cod.to_excel(r'C:\Users\estoque\Documents\estoque.renan\projetos\projeto_curva_abc\analise_curva_abc\divisao\verificacao_modulo\enderecos_modulo_29_28.xlsx')
```
