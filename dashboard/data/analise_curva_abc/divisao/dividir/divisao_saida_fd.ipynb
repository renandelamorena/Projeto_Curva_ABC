```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
```

```python
def refatorar_indece(df, nome_index):
    
    qnt_linha = df.shape[0] + 1

    index = [i for i in range(1, qnt_linha)]

    df.set_index(pd.Index(index), inplace=True)

    df.index.name = nome_index
    
    return df
```

```python
curva_abc_frac = pd.read_excel('dados_tratados\curva_abc_frac.xlsx')
curva_abc_frac.set_index('Ordem', inplace = True)
```

```python
selecao = curva_abc_frac['Descrição'].str.contains('fd', case=False)
fd = curva_abc_frac[selecao]
fd
```

```python
refatorar_indece(fd, 'Ordem')

cod_fd = fd['Código']

desc_fd = fd['Descrição']

saida_fd = fd['Qtde Venda Frac']
```

```python
corredor_29 = fd[['Código']].iloc[:AA]
refatorar_indece(corredor_29, None)
corredor_29.rename(columns = {'Código' : 'corredor_29'}, inplace = True)
corredor_29
```

```python
corredor_28 = fd[['Código']].iloc[(AA):(AA + AB)]
refatorar_indece(corredor_28, None)
corredor_28.rename(columns = {'Código' : 'corredor_28'}, inplace = True)
corredor_28
```

```python
corredor_29_outro_lado = fd[['Código']].iloc[(AA + AB):(AA + AB + 48)]
refatorar_indece(corredor_29_outro_lado, None)
corredor_29_outro_lado.rename(columns = {'Código' : 'corredor_29_outro_lado'}, inplace = True)
corredor_29_outro_lado
```

```python
pd.concat([corredor_29, corredor_28, corredor_29_outro_lado])
```

```python
junto = pd.merge(corredor_29, corredor_28, left_index=True, right_index=True)
junto = junto.merge(corredor_29_outro_lado, left_index=True, right_index=True)
junto
```
