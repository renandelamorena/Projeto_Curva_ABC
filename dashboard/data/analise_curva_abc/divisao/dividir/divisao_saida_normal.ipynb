## Funções

```python
def refatorar_indece(df, nome_index):
    
    qnt_linha = df.shape[0] + 1

    index = [i for i in range(1, qnt_linha)]

    df.set_index(pd.Index(index), inplace=True)

    df.index.name = nome_index
    
    return df
```

```python
modulos = ['29_28', '27_26', '25_24', '23_22', '21_20', '19_18']

def verificar_classificação():
    
    for modulo in modulos:
        # Transforma a primeira coluna em uma data frame diferente
        m_modulo = todos_classificar[[modulo]]

        # Junta o df da divisão, com o df de endereços atuais, ficando o df da divisão com os endereços de fracionados atuais
        m_modulo = pd.merge(m_modulo, end_atuais, left_on = modulo, right_on = 'Código') #1
        m_modulo.drop_duplicates(subset = [modulo], inplace = True)

        refatorar_indece(m_modulo, 'Ordem_modulo')

        # Verifica se esta no modulo correto
        m_modulo['Ender.Fracionado'] = m_modulo['Ender.Fracionado'].astype(str)
        m_modulo['Modulo correto'] = m_modulo['Ender.Fracionado'].str.startswith(modulos_corredor[modulo])

        # Adicionado columa com as classes
        m_modulo['Classes'] = index

        # Renomear colunas
        novos_nomes = {'Ender.Fracionado' : 'Endereço Atual',
                       'Código' : f'Itens do modulo {modulo}',
                      }
        m_modulo.rename(columns = novos_nomes, inplace = True)

        display(m_modulo)
    
verificar_classificação()
```
