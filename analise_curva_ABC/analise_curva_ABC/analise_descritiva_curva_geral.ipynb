# Funções

```python
def refatorar_indece(df, nome_index):
    
    qnt_linha = df.shape[0] + 1

    index = [i for i in range(1, qnt_linha)]

    df.set_index(pd.Index(index), inplace=True)

    df.index.name = nome_index
    
    return df
```

# Alta saída sem endereço

### Alta saída sem apanha de caixa fechada

```python
selecao_end_cx_nan = curva_geral['Apanha Cx'] == 'Não'

produtos_sem_apanha_cx = curva_geral[selecao_end_cx_nan].sort_values(by = 'Qtde Venda Geral', ascending = False).dropna(how='all')

colunas_irrelevantes_cx = ['Estoque Armaz. (Unid.)',
                           'Capacid.Chão',
                           'Estoque Chão',
                           'Capacidade Cx',
                           'Estoque Cx',
                           'Apanha Frac',
                           'Ender.Fracionado',
                           'Tipo',
                           'Capacidade Frac',
                           'Estoque Frac',
                           'Ender.Cx.Fechada',
                           'Apanha Cx',
                           'Estoque Geral (Unid.)'
                          ]
produtos_sem_apanha_cx = produtos_sem_apanha_cx.drop(colunas_irrelevantes_cx, axis = 1)

produtos_sem_apanha_cx.to_excel('analises_prontas\sem_enderecos\Produtos sem apanha cx fech.xlsx', index = False)
```

### Alta saída sem apanha fracionada

```python
selecao_end_frac_nan = curva_geral['Apanha Frac'] == 'Não'

produtos_sem_apanha_frac = curva_geral[selecao_end_frac_nan].sort_values(by = 'Qtde Venda Geral', ascending = False).dropna(how='all')

colunas_irrelevantes_frac = ['Capacid.Chão',
                             'Estoque Chão',
                             'Estoque Armaz. (Unid.)',
                             'Estoque Geral (Unid.)',
                             'Apanha Cx',
                             'Ender.Cx.Fechada',
                             'Capacidade Cx',
                             'Estoque Cx',
                             'Ender.Fracionado',
                             'Tipo',
                             'Capacidade Frac',
                             'Estoque Frac',
                             'Apanha Frac'
                          ]
produtos_sem_apanha_frac = produtos_sem_apanha_frac.drop(colunas_irrelevantes_frac, axis = 1)

produtos_sem_apanha_frac.to_excel('analises_prontas\sem_enderecos\Produtos sem apanha fracionada.xlsx', index = False)
```
