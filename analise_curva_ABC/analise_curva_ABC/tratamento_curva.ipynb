# Importando Bibliotecas

```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

# Importando Dados

```python
produtos = pd.read_excel('datasets\produtos.xlsx')
```

```python
curva_geral = pd.read_excel('datasets\curva_geral.xlsx')
```

# Tratando Dados

# Funções

```python
def tratar_codigos(dfs):
    
    for df in dfs:
        
        df['Cód.Produto'] = df['Cód.Produto'].astype(str)

        df['Cód.Produto'] = df['Cód.Produto'].str.slice(start=2)
        df['Cód.Produto'] = df['Cód.Produto'].str.lstrip('0')

    return df
```

## Tratamento Cadastro

```python
# exclui colunas irrelevantes para analise
colunas_irrelevantes = ['Qtde em Receb.',
                        'Estoque Armaz.',
                        'Descrição',
                        'Embal..1',
                       ]
produtos = produtos.drop(colunas_irrelevantes, axis = 1)

# renomeia coluna de cod
renomear_produtos = {'Código':'Cód.Produto',
                     'Apanha' : 'Apanha Cx',
                     'Apanha.1' : 'Apanha Frac',
                     'Capacidade' : 'Capacidade Cx',
                     'Capacidade.1' : 'Capacidade Frac',
                     'Estoque.1' : 'Estoque Frac',
                     'Estoque' : 'Estoque Cx',
                    }
produtos.rename(columns = renomear_produtos, inplace = True)

# exclui linhas completamente vazias
produtos = produtos.dropna(how='all')
```

 ## Tratamento Curvas

```python
# exclui colunas irrelevantes para analise
colunas_irrelevantes_curva = ['Qtde Acumulada',
                              'Perc.',
                              'Perc.Capacid.Apan.',
                              'Perc.Acumulada',
                              'Embal.',
                              'Endereço Apanha',
                              'Tipo Ender.',
                             ]

curva_geral = curva_geral.drop(colunas_irrelevantes_curva, axis = 1)

# exclui linhas completamente vazias
curva_geral = curva_geral.dropna(how='all')
```

## Tratanndo códigos

```python
dfs = [curva_geral, produtos]
tratar_codigos(dfs)
```

### Curva Geral

```python
renomear_geral = {'Estoque (Unid.)' : 'Estoque Chão',
                    'Ativ.Ressupr.' : 'Ativ.Ressupr.Frac',
                    'Curva' : 'Curva Geral',
                    'Qtde Venda' : 'Qtde Venda Geral',
                    'Dias Pedido' : 'Dias Pedido Geral',
                    'Média por dia' : 'Média por dia Geral',
                    'Ativ.Ressupr.' : 'Ativ.Ressupr. Geral',
                    'Capacid.Apan.' : 'Capacid.Chão'
                   }
curva_geral.rename(columns = renomear_geral, inplace = True)
```

```python
curva_geral
```

## Juntando as informações

```python
curva_abc_geral = pd.merge(curva_geral, produtos, on='Cód.Produto', how='outer')

curva_abc_geral['Cód.Produto'] = pd.to_numeric(curva_abc_geral['Cód.Produto'])

curva_abc_geral['Ender.Fracionado'] = curva_abc_geral['Ender.Fracionado'].astype(str)

curva_abc_geral
```

```python
qnt_linha = curva_abc_geral.shape[0] + 1

index = [i for i in range(1, qnt_linha)]

curva_abc_geral.set_index(pd.Index(index), inplace=True)

curva_abc_geral.index.name = 'Ordem'
curva_abc_geral
```

```python
curva_abc_geral.to_excel('dados_tratados\curva_abc_geral.xlsx')
```
